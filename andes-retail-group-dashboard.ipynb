# 📊 Proyecto: Dashboard de desempeño comercial 2024–2025
Empresa: **Andes Retail Group**

## 🔹 Paso 1: Conexión y Exploración

El dataset `Andes_Retail_Group_2024_2025.xlsx` fue importado en Tableau Public.

**Estructura del dataset:**
- 5,000 filas × 12 columnas
- Cada fila representa un pedido individual
- Período: 01/01/2024 al 31/12/2025
- Países: Perú, Chile, Colombia
- Categorías: Deportes, Electrónica, Hogar, Ropa
- Segmentos: Premium, Estándar, Económico

**Validación inicial:**
- Sin valores nulos en ninguna columna
- Sin filas duplicadas
- Ingresos = Precio_Unitario × Unidades_Vendidas
- Sin valores negativos en Ingresos ni Costo

## 🔹 Paso 2: Preparación de los datos

**Transformaciones realizadas en Tableau:**

- `Fecha_Pedido` → formato dd/MM/yyyy (español Latinoamérica)
- `ID_Pedido` → tipo Cadena (no es una medida numérica)
- Tipos de datos numéricos validados y correctos

**Campos calculados creados:**

`Ganancia` = [Ingresos] - [Costo]

`Margen %` = ([Ingresos] - [Costo]) / [Ingresos] → agregación: Promedio

`Nivel_Venta` = IF [Ingresos] >= 1000 THEN "Venta Alta" ELSE "Venta Baja" END

- **KPIs seleccionados y justificación**

    - Ingresos totales: Representa el total monetario generado por el negocio antes de descontar costos. Es el indicador más directo del volumen comercial.
    - Ganancia: Representa el resultado neto después de descontar costos. Permite evaluar si el negocio es financieramente sostenible.
    - Margen %: Expresa la ganancia como porcentaje sobre los ingresos. Permite evaluar la eficiencia de cada venta independientemente del volumen.
    - Total de pedidos: Indica el volumen de transacciones realizadas. Complementa los KPIs financieros al reflejar la actividad operativa del negocio.
- **Elección y justificación de gráficos por vista**

*Dashboard Overview*

- Gráfico de líneas — Ingresos mensuales 2024 vs 2025: Permite visualizar la evolución temporal del ingreso y comparar el desempeño entre ambos años mes a mes.
- Mapa — Ingresos por país: Presenta la distribución geográfica de los ingresos de forma visual e intuitiva, facilitando la identificación de mercados clave.
- Gráfico de barras — Ingresos por segmento de cliente: Permite comparar el aporte de cada segmento al ingreso total y establecer un ranking claro entre ellos.

*Dashboard Detalle*

####Gráficos
- Gráfico de barras — Ingresos por estación: Compara el total de ingresos por temporada. Revela la brecha crítica entre Verano ($2.2M) e Invierno ($653K).

- Heatmap — Categoría × Estación: Cruza dos dimensiones para identificar combinaciones específicas de bajo desempeño. Revela que Ropa no se recupera en primavera y Electrónica cae más en invierno.
- Tabla — Ingresos, Ganancia, Pedidos y Clientes únicos por Categoría: Complementa los gráficos con métricas de volumen que no aparecen visualmente en otros gráficos.

## 🔹 Paso 5: Narrativa SCQA

### Dashboard Overview

**S (Situación):**
Andes Retail Group opera en Colombia, Chile y Perú comercializando productos de Deportes, Electrónica, Hogar y Ropa durante 2024 y 2025. En ese período generó \$5.5M en ingresos con un margen promedio del 35% y 5,000 pedidos.

**C (Complicación):**
Sin embargo, el análisis revela tres tensiones: Colombia concentra solo el 24% de los ingresos frente al liderazgo de Perú, el segmento Económico aporta apenas el 9% del total, y los ingresos caen drásticamente en invierno, generando solo \$653K frente a \$2.2M en verano.

**Q (Pregunta):**
¿Dónde están las oportunidades de mejora y qué las explica?

**A (Respuesta):**
El análisis de detalle revela que la caída estacional afecta a todas las categorías en invierno, con Electrónica como la más impactada. Ropa y Hogar quedan consistentemente por debajo del promedio de ingresos. Se recomienda: (1) investigar la penetración de mercado en Colombia, (2) evaluar la viabilidad de escalar el segmento Estándar en lugar del Económico, y (3) diseñar campañas comerciales focalizadas en invierno, especialmente para Electrónica y Ropa.

---

### Dashboard Detalle

**S (Situación):**
El análisis detallado cruza categorías de producto, estaciones y segmentos para identificar patrones específicos de bajo desempeño en el período 2024-2025.

**C (Complicación):**
Todas las categorías caen en invierno, pero Electrónica es la más afectada con \$153K. Ropa no se recupera en primavera (\$285K), siendo la categoría más débil en ese período. Los 3 segmentos tienen el mismo margen del 35%, lo que indica que las diferencias entre ellos son de volumen, no de eficiencia.

**Q (Pregunta):**
¿Qué combinaciones de categoría y estación representan las mayores oportunidades de intervención comercial?

**A (Respuesta):**
Electrónica en invierno y Ropa en primavera son las combinaciones de menor desempeño. Dado que el margen es igual en todos los segmentos, la palanca de crecimiento está en aumentar el volumen de ventas, no en mejorar la eficiencia. Se recomienda activar promociones específicas para Electrónica en invierno y Ropa en primavera, priorizando los mercados con mayor base de clientes.
